```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
topk_jeep_path = "..\\trained_gans\\projgan-topk-jeep\\"
topk5_fish_path = "..\\trained_gans\\projgan-topk32_5-fish\\"
topk20_fish_path = "..\\trained_gans\\projgan-topk32_20-fish\\"
topk50_fish_path = "..\\trained_gans\\projgan-topk32_50-fish\\"
regular_jeep_path = "..\\trained_gans\\projgan-regular-jeep\\"
regular_fish_path = "..\\trained_gans\\projgan-regular-fish\\"
```

```python
topk_json = pd.read_json(path_or_buf=topk_jeep_path+"stats.jsonl", lines=True)
topk_json.head()
```

```python
normal_json = pd.read_json(path_or_buf=regular_jeep_path+"stats.jsonl", lines=True)
normal_json.head()
```

```python
normal_G_loss = [d.get('mean') for d in normal_json['Loss/G/loss']]
plt.plot(np.arange(0, len(normal_G_loss)/2, 0.5), normal_G_loss, label="Top-K")
topk_G_loss = [d.get('mean') for d in topk_json['Loss/G/loss']]
plt.plot(np.arange(0, len(topk_G_loss)/2, 0.5), topk_G_loss, label="Regular")
plt.legend()
plt.xlabel("kimg")
plt.ylabel("Loss")
plt.title("Generator Losses")
plt.show()

topk_D_loss = [d.get('mean') for d in topk_json['Loss/D/loss']]
plt.plot(np.arange(0, len(topk_D_loss)/2, 0.5), topk_D_loss, label="Top-K")
normal_D_loss = [d.get('mean') for d in normal_json['Loss/D/loss']]
plt.plot(np.arange(0, len(normal_D_loss)/2, 0.5), normal_D_loss, label="Regular")
plt.legend()
plt.xlabel("kimg")
plt.ylabel("Loss")
plt.title("Discriminator Losses")
plt.show()
```

```python
topk_kid_jeep = pd.read_json(path_or_buf=topk_jeep_path+"metric-kid50k_full.jsonl", lines=True)
normal_kid_jeep = pd.read_json(path_or_buf=regular_jeep_path+"metric-kid50k_full.jsonl", lines=True)
topk5_kid_fish = pd.read_json(path_or_buf=topk5_fish_path+"metric-kid50k_full.jsonl", lines=True)
topk20_kid_fish = pd.read_json(path_or_buf=topk20_fish_path+"metric-kid50k_full.jsonl", lines=True)
topk50_kid_fish = pd.read_json(path_or_buf=topk50_fish_path+"metric-kid50k_full.jsonl", lines=True)
normal_kid_fish = pd.read_json(path_or_buf=regular_fish_path+"metric-kid50k_full.jsonl", lines=True)
```

```python
topk_kid_mean = [d.get('kid50k_full')*1000 for d in topk_kid_jeep['results']]
plt.plot(np.arange(0, len(topk_kid_mean)), topk_kid_mean, label="Top-K Jeep")
normal_kid_mean = [d.get('kid50k_full')*1000 for d in normal_kid_jeep['results']]
plt.plot(np.arange(0, len(normal_kid_mean)), normal_kid_mean, label="Regular Jeep")
plt.xlabel("Evaluation Step")
plt.ylabel("KID")
plt.title("KID Scores for GANs")
plt.legend()
plt.show()

topk_kid_mean = [d.get('kid50k_full')*1000 for d in topk5_kid_fish['results']]
plt.plot(np.arange(0, len(topk_kid_mean)), topk_kid_mean, label=r"Top-K 5% Fish")
topk_kid_mean = [d.get('kid50k_full')*1000 for d in topk20_kid_fish['results']]
plt.plot(np.arange(0, len(topk_kid_mean)), topk_kid_mean, label=r"Top-K 20% Fish")
topk_kid_mean = [d.get('kid50k_full')*1000 for d in topk50_kid_fish['results']]
plt.plot(np.arange(0, len(topk_kid_mean)), topk_kid_mean, label=r"Top-K 50% Fish")
normal_kid_mean = [d.get('kid50k_full')*1000 for d in normal_kid_fish['results']]
plt.plot(np.arange(0, len(normal_kid_mean)), normal_kid_mean, label="Regular Fish")
plt.xlabel("Evaluation Step")
plt.ylabel("KID")
plt.title("KID Scores for GANs")
plt.legend()
plt.show()
```
